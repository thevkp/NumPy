
# üß† Sigmoid Function: Introduction & Uses

---

## üìå What is the Sigmoid Function?

The **sigmoid function** is a mathematical function that maps any real number into a value between **0 and 1**.

### üî£ Formula:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

This S-shaped (sigmoid) curve is smooth and differentiable, making it a common choice in various computational fields.

---

## üìà Why Use Sigmoid?

### ‚úÖ Key Properties:

- Output is **bounded** between 0 and 1.
- **Differentiable** everywhere.
- **Monotonic** ‚Äî as input increases, output increases.
- Has a simple **derivative**:

  $$
  \sigma'(x) = \sigma(x)(1 - \sigma(x))
  $$

---

## ‚öôÔ∏è Where Is It Used?

### 1. **Logistic Regression**
- Used to model probabilities.
- Output represents the **probability** of a binary class.

### 2. **Neural Networks (as Activation Function)**
- Introduces **non-linearity** in hidden layers.
- Smooth gradient helps in **gradient-based optimization**.
- Historically used in early neural nets.

### 3. **Binary Classification**
- Converts linear scores to probabilities between 0 and 1.

---

## üíª Python Example

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("œÉ(x)")
plt.grid(True)
plt.show()
```

---

## ‚ö†Ô∏è Limitations

- **Vanishing gradient** for large |x| (saturates at 0 or 1).
- Not zero-centered ‚Üí can slow training in deep networks.
- Often replaced by ReLU in modern deep learning architectures.

---

## ‚úÖ Summary

| Feature        | Description                      |
|----------------|----------------------------------|
| Range          | 0 to 1                           |
| Differentiable | Yes                              |
| Smooth         | Yes                              |
| Use Case       | Logistic regression, old NNs     |

Sigmoid helps model **probabilities**, and its simplicity made it a classic ‚Äî though it's been partially replaced by more modern functions like ReLU and softmax.


# üßÆ Mean Squared Error (MSE) with NumPy

---

## üìå What is Mean Squared Error?

The **Mean Squared Error (MSE)** is a metric used to measure the average of the **squares of differences** between corresponding elements in two arrays.

> Think of it as a way to measure how "off" your predictions (or approximations) are from the actual values.

---

## üß† Formula:

Given two arrays:  
- `y_true = [y‚ÇÅ, y‚ÇÇ, ..., y‚Çô]`  
- `y_pred = [≈∑‚ÇÅ, ≈∑‚ÇÇ, ..., ≈∑‚Çô]`

The formula is:

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

---

## üíª NumPy Code Example

```python
import numpy as np

# Two example vectors
y_true = np.array([3, -0.5, 2, 7])
y_pred = np.array([2.5, 0.0, 2, 8])

# Compute MSE
mse = np.mean((y_true - y_pred) ** 2)
print("Mean Squared Error:", mse)
```

üßæ Output:

```
Mean Squared Error: 0.375
```

---

## üîç Why Squared?

- Squaring ensures **positive values**, so negative differences don't cancel out.
- Emphasizes **larger errors** more than smaller ones.
- Makes the function smooth and differentiable.

---

## ‚öôÔ∏è Applications Outside AI/ML

Even if you're not into AI, MSE is still useful:
- üìê Approximating noisy signals
- üéÆ Measuring control system errors
- üìä Comparing predictions vs measurements
- üé® Graphics: comparing pixel intensity values

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
actual = np.random.randint(1, 50, 25)
predicted = np.random.randint(1, 50, 25)

# Compute MSE
MSE = np.mean((actual - predicted) ** 2)
print("Actual:", actual)
print("Predicted:", predicted)
print("MSE:", MSE)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(actual, predicted, color='blue', label='Predicted Points')
plt.plot([0, 50], [0, 50], color='red', linestyle='--', label='Perfect Prediction Line (y = x)')
plt.title(f'Actual vs Predicted\nMSE = {MSE:.2f}')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.grid(True)
plt.xlim(0, 55)
plt.ylim(0, 55)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

actual_price = np.array([250, 300, 150, 400])
predicted_price = np.array([270, 290, 180, 370])

MSE = np.mean((actual_price - predicted_price) ** 2)
print("MSE:", MSE)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(actual_price, predicted_price, color='blue', label='Predicted Points')

# Fix the line to cover full range of prices
min_val = min(actual_price.min(), predicted_price.min()) - 20
max_val = max(actual_price.max(), predicted_price.max()) + 20

plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='Perfect Prediction Line (y = x)')

plt.title(f'Actual vs Predicted\nMSE = {MSE:.2f}')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.legend()
plt.grid(True)
plt.xlim(min_val, max_val)
plt.ylim(min_val, max_val)
plt.show()
3
````

````python
# Working with missing values
a = np.array([1,2,3,4,np.nan,6])
b = np.array([2,1,4,np.nan,5,6])

mse = np.mean((a - b)  ** 2)
print(mse)

#Boolean indexing to deal with nan values
#isnan(n) returns only the nan value i.e. True if nan
boolean_mask_a = np.isnan(a)
print(boolean_mask_a)

filtered_a = a[boolean_mask_a]
print(filtered_a)

boolean_mask_b = np.isnan(b)
print(boolean_mask_b)

filtered_b = b[boolean_mask_b]
print(filtered_b)
````

````python
#True where both 'a' and 'b' are nan
mask = ~(np.isnan(a)) & ~(np.isnan(b))

filtered_a = a[mask]
filtered_b = b[mask]

print(filtered_a)
print(filtered_b)

mse = np.mean((a - b) ** 2)
print(mse)

mse = np.mean((filtered_a - filtered_b) ** 2)
print(mse)
````

````python
#plot a circle centered at origin with radius 5
import numpy as np
import matplotlib.pyplot as plt

r = 5
x = np.linspace(-r, r, 1000)

# Upper and lower semicircles
y_upper = np.sqrt(r**2 - x**2)
y_lower = -np.sqrt(r**2 - x**2)

plt.figure(figsize=(6,6))
plt.plot(x, y_upper, color='blue')
plt.plot(x, y_lower, color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title(r'Circle: $x^2 + y^2 = 25$')
plt.gca().set_aspect('equal')  # ensure circle doesn't look squashed
plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 1000)
r = 5
h, k = 2, 3  # center of the circle

x = h + r * np.cos(theta)
y = k + r * np.sin(theta)

plt.figure(figsize=(6,6))
plt.plot(x, y, label='Circle')
plt.scatter(h, k, color='red', label='Center')
plt.gca().set_aspect('equal')  # make it look like a circle
plt.grid(True)
plt.title("Circle: (x - 2)¬≤ + (y - 3)¬≤ = 25")
plt.legend()
plt.show()
````
